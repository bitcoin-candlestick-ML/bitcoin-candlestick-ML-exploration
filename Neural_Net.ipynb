```python
import pandas as pd
import seaborn as sns
import numpy as np

pd.options.display.max_rows = 100
pd.options.display.max_columns = 100

raw = pd.read_csv ('21_features.csv',parse_dates=['Date'])
raw
```

```python
# convert index to column so that we can reference by event number instead of relying on index
# run only once! else restart kernal

raw.reset_index(inplace=True)
raw = raw.rename(columns = {'index':'event_number'})
```

```python
raw.head()
```

```python
df = raw[['event_number','A','B','C','D','E','F','AB_Days','BC_Days','CD_Days','DE_Days','AB_Vert','BC_Vert','CD_Vert','DE_Vert','AB','BC','CD','DE','EF']]
```

```python
df.head()
```

```python
df = df.rename(columns = {'AB':'AB_avg_change_per_day','BC':'BC_avg_change_per_day','CD':'CD_avg_change_per_day','DE':'DE_avg_change_per_day','EF':'EF_avg_change_per_day'})
```

```python
df.head()
```

```python
df['AB_percent_change_per_day']= df['AB_avg_change_per_day']/df['A']
df['BC_percent_change_per_day']= df['BC_avg_change_per_day']/df['B']
df['CD_percent_change_per_day']= df['CD_avg_change_per_day']/df['C']
df['DE_percent_change_per_day']= df['DE_avg_change_per_day']/df['D']
df['EF_percent_change_per_day']= df['EF_avg_change_per_day']/df['E']
```

```python
df['bs_1_or_sb_0']=np.where((df['F']>df['E']),'1','0')
```

```python
df.head()
```

```python
df['AB_Days'].value_counts()
```

```python
# our train set should be 80-85% of the number of inputs. Let's start with 80%:
21*.8
```

```python
df.columns
```

```python
nn_features = df[['event_number', 'A', 'B', 'C', 'D', 'E', 'AB_Days', 'BC_Days',
       'CD_Days', 'DE_Days', 'AB_Vert', 'BC_Vert', 'CD_Vert', 'DE_Vert',
       'AB_avg_change_per_day', 'BC_avg_change_per_day',
       'CD_avg_change_per_day', 'DE_avg_change_per_day', 'AB_percent_change_per_day',
       'BC_percent_change_per_day', 'CD_percent_change_per_day',
       'DE_percent_change_per_day', 'bs_1_or_sb_0']]
```

```python
# time to train our model!

from sklearn.model_selection import train_test_split

train_data, test_data = np.split(nn_features, [int(.80*len(df))])

test = test_data
train = train_data
target = 'bs_1_or_sb_0'
train, val = train_test_split(train, test_size=0.2,  
                              stratify=train[target])
features = train.columns

X_train = train.drop(columns=target)
y_train = train[target]
X_test = test.drop(columns=target)
y_test = test[target]
X_val = val.drop(columns=target)
y_val = val[target]
print(features)
```

```python
encoder = ce.OneHotEncoder(use_cat_names=True)
X_val = encoder.fit_transform(X_val)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

for k in range(1, len(X_train.columns) + 1):
  print(f'{k} features')

  selector = SelectKBest(k=k)
  X_train_selected = selector.fit_transform(X_train, y_train)
  X_test_selected = selector.transform(X_val)

  model = LinearRegression()
  model.fit(X_train_selected, y_train)
  y_pred = model.predict(X_test_selected)
  mae = mean_absolute_error(y_val, y_pred)
  print(f'Test MAE: {mae:,.4f} \n')
```

```python
from IPython.display import display, HTML
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge

for alpha in [0.00000000001, 0.00001, 0.001, .1]:
    
    # Fit Ridge Regression model
    display(HTML(f'Ridge Regression, with alpha={alpha}'))
    model = Ridge(alpha=alpha, normalize=True)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_val)

    # Get Test MAE
    mae = mean_absolute_error(y_val, y_pred)
    display(HTML(f'Test Mean Absolute Error: {mae:,.4f}'))
    
    # Plot coefficients
    coefficients = pd.Series(model.coef_, X_train.columns)
    plt.figure(figsize=(16,8))
    coefficients.sort_values().plot.barh(color='blue')
    plt.xlim(-5,5)
    plt.show()
```

```python
# alpha at .1 seems to be the lowest MAE that I tested and the other alphas 
# don't go any lower than MAE of .5081 so we'll go with that
linridge = Ridge(alpha=0.1).fit(X_train, y_train)

print('R-squared score (train): {:.3f}'
     .format(linridge.score(X_train, y_train)))
```

```python
linridge = Ridge(alpha=0.1).fit(X_val, y_val)

print('R-squared score (validate): {:.3f}'
     .format(linridge.score(X_val, y_val)))
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import cross_val_score

pipeline = make_pipeline(
    ce.OneHotEncoder(use_cat_names=True), 
    SimpleImputer(strategy='mean'), 
    StandardScaler(), 
    SelectKBest(k=18), 
    Ridge(alpha=0.1)
)

k = 3
scoresLM = cross_val_score(pipeline, X_train, y_train, cv=k, 
                         scoring='neg_mean_absolute_error')
print(f'MAE for {k} folds:', -scoresLM)
```
